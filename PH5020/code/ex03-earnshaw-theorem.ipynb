```python
import numpy as np
import matplotlib.pyplot as plt
```

## Harmonic potential

```python
def system_potential_spring(x, y):
    V  = x**2 + y**2
    return V
```

```python
L=2
x1, x2 = -L, L
y1, y2 = -L, L

X, Y = np.meshgrid(np.linspace(x1, x2, 128), np.linspace(y1, y2, 128))


V=system_potential_spring(X, Y)
Fy, Fx = np.gradient(-V)

fig = plt.figure(num=None, figsize=(13, 6), dpi=80, facecolor='w', edgecolor='k')

plt.subplot(121)
CS=plt.contour(X, Y, V, 16, colors='grey', linewidths=1.4, linestyles='dotted')
plt.contourf(X,Y, V, 8,cmap=plt.cm.GnBu)#, norm=colors.PowerNorm(gamma=0.5))
plt.axis('equal')

plt.subplot(122)
CS=plt.contour(X, Y, V, 16, colors='grey', linewidths=1.4, linestyles='dotted')
plt.contourf(X,Y, V, 8,cmap=plt.cm.GnBu)#, norm=colors.PowerNorm(gamma=0.5))
plt.streamplot(X,Y, Fx, Fy, density=0.5, linewidth=1,  color='indianred', broken_streamlines=False, arrowsize=2);
plt.axis('equal');
```

## Coulomb potential

We now consider the Coulomb potential in 2D and 3D due to point charges. 

```python
nu=0.5

def system_potential(x, y):
    dx = x - 0; dy = y -0 
    V = 1/(dx**2 + dy**2)**(nu)
    
    dx = x - 2; dy = y -0 
    V += 1/(dx**2 + dy**2)**(nu)

    dx = x - 2; dy = y - 2 
    V += 1/(dx**2 + dy**2)**(nu)

    dx = x - 0; dy = y -2 
    V += 1/(dx**2 + dy**2)**(nu)
    return V

    
def system_potential2D(x, y):
    dx = x - 0; dy = y -0 
    V = np.log((dx**2 + dy**2)**(nu))
    
    dx = x - 2; dy = y -0 
    V += np.log((dx**2 + dy**2)**(nu))

    dx = x - 2; dy = y - 2 
    V += np.log((dx**2 + dy**2)**(nu))

    dx = x - 0; dy = y -2 
    V += np.log((dx**2 + dy**2)**(nu))
    return -V
```

```python
del_x = 0.5
L=2
x1, x2 = del_x, L-del_x
y1, y2 = del_x, L-del_x

X, Y = np.meshgrid(np.linspace(x1, x2, 128), np.linspace(y1, y2, 128))
```

## 3D potential

```python
V=system_potential(X, Y)
Fy, Fx = np.gradient(-V)

fig = plt.figure(num=None, figsize=(13, 6), dpi=80, facecolor='w', edgecolor='k')

plt.subplot(121)
CS=plt.contour(X, Y, V, 32, colors='grey', linewidths=1.4, linestyles='dotted')
plt.contourf(X,Y, V, 8,cmap=plt.cm.GnBu)#, norm=colors.PowerNorm(gamma=0.5))

plt.subplot(122)
CS=plt.contour(X, Y, V, 32, colors='grey', linewidths=1.4, linestyles='dotted')
plt.contourf(X,Y, V, 8,cmap=plt.cm.GnBu)#, norm=colors.PowerNorm(gamma=0.5))
plt.streamplot(X,Y, Fx, Fy, density=0.5, linewidth=1,  color='indianred', broken_streamlines=False, arrowsize=2);
```

It looks like the centre is a stable point. But it is unstable in a direction perpendicular to the plane. 

## 2D Potential

```python
V=system_potential2D(X, Y)
Fy, Fx = np.gradient(-V)

fig = plt.figure(num=None, figsize=(13, 6), dpi=80, facecolor='w', edgecolor='k')

plt.subplot(121)
CS=plt.contour(X, Y, V, 64, colors='grey', linewidths=1.4, linestyles='dotted')
plt.contourf(X,Y, V, 8,cmap=plt.cm.GnBu)#, norm=colors.PowerNorm(gamma=0.5))

plt.subplot(122)
CS=plt.contour(X, Y, V, 64, colors='grey', linewidths=1.4, linestyles='dotted')
plt.contourf(X,Y, V, 8,cmap=plt.cm.GnBu)#, norm=colors.PowerNorm(gamma=0.5))
plt.streamplot(X,Y, Fx, Fy, density=0.5, linewidth=1,  color='indianred', broken_streamlines=False, arrowsize=2);
```
