```python
import numpy as np
import matplotlib.pyplot as plt
```

## Point charge near a grounded conducting plane

We now consider the potential due to a point charge near a grounded plane

```python
nu=0.5
Ng=128

h = 1

def system_potential_Free(x, y):
    V = np.zeros((Ng, Ng))
    for i in range(Ng):
        for j in range(Ng):
            dx = x[i,j] - 0; dy = y[i,j] -h 
            V[i,j] = 1/(dx**2 + dy**2)**(nu)
    
            dy = y[i,j] +h 
            V[i,j] -= 1/(dx**2 + dy**2)**(nu)
    return V
```

```python
L=2
x1, x2 = -L, L
y1, y2 = 0, L

X, Y = np.meshgrid(np.linspace(x1, x2, Ng), np.linspace(y1, y2, Ng))
```

```python
V=system_potential_Free(X, Y)
Fy, Fx = np.gradient(-V)

fig = plt.figure(num=None, figsize=(24, 6), dpi=80, facecolor='w', edgecolor='k')

plt.subplot(121)
CS=plt.contour(X, Y, V, 256, colors='grey', linewidths=1.4, linestyles='dotted')
plt.contourf(X,Y, V, 8,cmap=plt.cm.GnBu)#, norm=colors.PowerNorm(gamma=0.5))

plt.subplot(122)
CS=plt.contour(X, Y, V, 256, colors='grey', linewidths=1.4, linestyles='dotted')
plt.contourf(X,Y, V, 8,cmap=plt.cm.GnBu)#, norm=colors.PowerNorm(gamma=0.5))
plt.streamplot(X,Y, Fx, Fy, density=0.25, linewidth=1,  color='indianred', broken_streamlines=False, arrowsize=2);
```

## Point charge near a dielectric

We now consider the potential due to a point charge near a grounded plane

```python
nu=0.5
Ng=128

h = 1


def system_potential(x, y,e1, e2 ):
    V = np.zeros((Ng, Ng))
    fac1=-(e2-e1)/(e1+e2)
    fac2=(2*e2)/(e1+e2)
    fac_e1 = 1/(4*np.pi*e1)
    fac_e2 = 1/(4*np.pi*e2)
    
    for i in range(Ng):
        for j in range(Ng):
            if y[i,j]>0:
                dx = x[i,j] - 0; dy = y[i,j] -h 
                V[i,j] = fac_e1/(dx**2 + dy**2)**(nu)
        
                dy = y[i,j] +h 
                V[i,j] += fac_e1*fac1/(dx**2 + dy**2)**(nu)
            else:
                
                dx = x[i,j] - 0; dy = y[i,j] -h 
                V[i,j] = fac_e2*fac2/(dx**2 + dy**2)**(nu)
    return V
```

```python
L=2


x1, x2 = -L, L
y1, y2 = -L, L

X, Y = np.meshgrid(np.linspace(x1, x2, Ng), np.linspace(y1, y2, Ng))
```

```python
e1, e2 = 1, 4
V=system_potential(X, Y, e1, e2)
Fy, Fx = np.gradient(-V)

fig = plt.figure(num=None, figsize=(13, 6), dpi=80, facecolor='w', edgecolor='k')

plt.subplot(121)
CS=plt.contour(X, Y, V, 256, colors='grey', linewidths=1.4, linestyles='dotted')
plt.contourf(X,Y, V, 8,cmap=plt.cm.GnBu)#, norm=colors.PowerNorm(gamma=0.5))

plt.subplot(122)
CS=plt.contour(X, Y, V, 256, colors='grey', linewidths=1.4, linestyles='dotted')
plt.contourf(X,Y, V, 8,cmap=plt.cm.GnBu)#, norm=colors.PowerNorm(gamma=0.5))
plt.streamplot(X,Y, Fx, Fy, density=0.25, linewidth=1,  color='indianred', broken_streamlines=False, arrowsize=2);
```
