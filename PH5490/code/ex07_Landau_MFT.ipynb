# Landau theory
> a phenomenological theory of phase transitions


```python
import numpy as np
from numpy.random import rand
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(num=None, figsize=(6, 6), dpi=100, facecolor='w', edgecolor='k')
plt.rcParams.update({'font.size': 16}); plt.grid()

def freeEnergy(a, b, m):
    return 0.5*a*m**2 + 0.25*b*m**4

Nt=64
m = np.linspace(-1.5, 1.5, Nt)
b = 1

a = 0;  plt.plot(m, freeEnergy(a, b, m), '-.', lw=2, label='$T=T_c$')
a = -1;  plt.plot(m, freeEnergy(a, b, m), '-', lw=2, label='$T<T_c$')
a = 1;  plt.plot(m, freeEnergy(a, b, m), ':', lw=2, label='$T>T_c$')
plt.xlabel(r'$m$'); plt.ylim(-0.5, 1.5)
plt.legend();
```

#### Minimizing the free energy

Minimising the free energy with respect to $m$, we get

$$
am + bm^3=0
$$

Thus, the solutions are:

$$
m=\pm\sqrt{\frac{-a}b},\qquad m=0
$$

The critical exponent is:

$$
m\propto (T_c-T)^{0.5}
$$

## Plotting the mean field solution


```python
fig = plt.figure(num=None, figsize=(8, 6), dpi=100, facecolor='w', edgecolor='k')
plt.rcParams.update({'font.size': 16}); plt.grid()

def orderParameter(T, Tc):
    N = np.size(T); m=np.zeros(N)
    for i in range(N):
        if T[i]<=Tc:   m[i] = (Tc-T[i])**0.5
        else:         m[i] = 0
    return m


Nt=256
T = np.linspace(0, 1.62, Nt)
Tc = 1

plt.plot(T, orderParameter(T, Tc), '.-', lw=4, label='$T=T_c$')
```

## The cubic term

The cubic term and line of first order phase transitions

$$
f=f_0 + \frac12 am^2 - \frac13 cm^3 +  \frac14 b m^4  
$$

```python
fig = plt.figure(num=None, figsize=(9, 5), dpi=128, facecolor='w', edgecolor='k')
plt.rcParams.update({'font.size': 18});



Nt=64
m = np.linspace(-1.5, 1.5, Nt)

b=7.6

def freeEnergy(a, b, m, c):
    return  0.5*a*m**2 + 0.25*b*m**4 - 1/3*c*m**3

a0=3
plt.plot(m, m*0, '-',c='grey', lw=1 )
a, c = a0, 0*np.sqrt(9/2);    plt.plot(m, freeEnergy(a, b, m, c), '-',color='k', lw=2, label=r'$c^2=0$')
a, c = a0, 1*np.sqrt(4*a*b);    plt.plot(m, freeEnergy(a, b, m, c), '.-',color='green', lw=2, label=r'$c^2=4ab$')
a, c = a0, 1*np.sqrt(9*a*b/2);    plt.plot(m, freeEnergy(a, b, m, c), '--',color='indianred', lw=2, label=r'$c^2=9ab/2$')
a, c = a0, 1*np.sqrt(5*a*b);    plt.plot(m, freeEnergy(a, b, m, c), ':',color='olive', lw=2, label=r'$c^2=5ab$')
plt.grid(); plt.yticks([], []);
plt.ylim(-.5,1);
plt.legend()
```
