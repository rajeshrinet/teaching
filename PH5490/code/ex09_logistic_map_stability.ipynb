## The logistic map

$$
x_{n+1}=r\, x_{n}(1-x_{n})=f(x_n),\qquad 0<x<1,\,\,\,r>1.
$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 32

x = np.zeros(n)
r = 2.8
x[0] = 0.2
```

```python
for i in range(n-1):
    x[i+1] = r*x[i]*(1-x[i])
```

```python
plt.plot(x, 'o-')
plt.xlabel('n')
plt.ylabel('x_n')
```

## Fixed points

At the fixed point

$$
x^*_{n+1}=x_n^*
$$

Thus, 

$$
x^*_n= r\,x^*_{n}(1-x^*_{n})
$$

We have fixed points at:

$$
x^*=0
$$

and 

$$
x^*=1-\frac1{r}
$$


```python
1- 1/r
```

```python
fig = plt.figure(num=None, figsize=(6, 6), dpi=100, facecolor='w', edgecolor='k')
plt.rcParams.update({'font.size': 12}); plt.grid()


Nt=64
xn = np.linspace(0, 1, Nt)
yn = r*xn*(1-xn)
plt.plot(xn, yn, color='indianred', lw=2, label='f(x_n)')
plt.plot(xn, xn, '--', color='midnightblue', lw=2, label='x_n')

xs = 1-1/r; ys=r/4
plt.plot(xs, xs, 'mo', ms=12, label='stable fixed point')
plt.plot(xs*0, xs*0, 'ko', ms=12, label='unstable fixed point')
plt.legend()
plt.xlabel('x_n')
plt.ylabel('x_n+1')
```

## Stability of the fixed points

Which fixed points are stable and which are unstable?

Consider the fixed point $x^*_n=x^*_{n+1}=x^*$

Assume a small perturbation 

$$
x_n = x^* + \delta_n
$$

$$
\delta_{n+1}= \Lambda(x^*)\,\delta_{n}.
$$

Here, we have

$$
\Lambda(x^*) = r\, (1-2x^*)
$$

A fixed point is :

* stable if $\Lambda(x^*)$ <1
* unstable if $\Lambda(x^*)$ >1
