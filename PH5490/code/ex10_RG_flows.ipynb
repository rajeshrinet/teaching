```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({
    "text.usetex": True,
    "font.family": "Times",
    "font.sans-serif": "Times",
})
```

```python
fig=plt.figure(figsize=(10,5))
x = np.linspace(0.8, 2, 64)
plt.plot(x, 2*x, label='2x')
plt.plot(x, x*x+1/(x*x), label=r'$x^2 + 1/(x^2)$')
plt.plot(x*0+1.84, 2*x, '--', label='x=1.84')
plt.legend(); plt.axis('tight')
plt.grid()
```

```python
fig = plt.figure(num=None, figsize=(6, 6), dpi=256, facecolor='w', edgecolor='k')
plt.rcParams.update({'font.size': 16}); plt.grid()

x = np.linspace(0.0, 0.6, 64)
plt.plot(x, x, label=r'$K$')
plt.plot(x, 0.5*np.log(np.cosh(2*x)), label=r'$0.5 \ln (\cosh 2K)$')
plt.plot(x, 0.5*np.log(np.cosh(4*x)), label=r'$0.5 \ln (\cosh 4K)$')
plt.xlabel(r'$K$')
plt.legend(); 
```

## Self-similarity of random walks

The overall shape of the random walk trajectory doesn't change at different scales.

```python
def RW(N):
    return np.cumsum(1-2*np.random.randint(0, 2, N))     # radom walk code
```

```python
fig = plt.figure(num=None, figsize=(20, 6), dpi=200, facecolor='w', edgecolor='k')


plt.subplot(231)
N=512; plt.plot(RW(N));   

plt.subplot(232)
N=1024; plt.plot(RW(N));   ## RW2


plt.subplot(233)
N=2048; plt.plot(RW(N));   ## RW3


plt.subplot(234)
N=4096; plt.plot(RW(N));   ## RW4


plt.subplot(235)
N=8000; plt.plot(RW(N));   ## RW5


plt.subplot(236)
N=20000; plt.plot(RW(N));   ## RW6
```
