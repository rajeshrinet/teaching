```python
import numpy as np
import matplotlib.pyplot as plt
```

## Flow due to 1s mode

$$
v_i = G_{ij} F_j
$$

$$
v_i =\frac{1}{8\pi\eta} \left(\frac{F_i}{r} + \frac{(F_j r_j)\,r_i}{r^3}\right)
$$

```python
xx=6;  x = np.linspace(-xx,xx,128); 
 
Lx, Ly = 2*np.pi, 2*np.pi
Np =64
xG, zG = np.meshgrid(np.linspace(-Lx, Lx, Np), np.linspace(-Lx, Lx, Np))

fx, fz = 0, -1

r = np.sqrt(xG*xG + zG*zG)
idr = 1/r
fD = (fx*xG + fz*zG)*idr*idr

vx = (fx + fD*xG)*idr
vz = (fz + fD*zG)*idr

plt.streamplot(xG, zG, vx, vz,density=0.8, arrowsize=2, broken_streamlines=True); 
spd = vx*vx + vz*vz; plt.pcolor(xG, zG, np.log(spd), cmap=plt.cm.bone); plt.colorbar()
plt.plot(0, 0, 'o', ms=32, mfc='w')
```

## Flow due to 2s mode

$$
v^{2s}_i =\tilde\nabla_k G_{ij}\, S_{jk}
$$

$$
v^{2s}_i = \frac{3 s_0}{8\pi\eta r^5} r_i r_j r_k S_{ij}
$$


We use $S_{jk} = s_0 \left(e_i e_j - \frac13\delta_{ij}\right)$. Thus, we get:


$$
v^{2s}_i = \frac{s_0}{8\pi\eta} \left( \frac{3(\boldsymbol e\cdot\boldsymbol r)^2}{r^5} r_i - \frac{e_i}{r^3}  \right)
$$


```python
xx=6;  x = np.linspace(-xx,xx,256); 
 
Lx, Ly = 2*np.pi, 2*np.pi
Np =64
xG, zG = np.meshgrid(np.linspace(-Lx, Lx, Np), np.linspace(-Lx, Lx, Np))
s0=1
r = np.sqrt(xG*xG + zG*zG)
idr = 1/r
idr2 = idr*idr 

ex, ez = 0, 1
eR2 = (ex*xG + ez*zG)*(ex*xG + ez*zG)*idr2

vx = s0*(3*eR2*xG - xG)*idr2*idr
vz = s0*(3*eR2*zG - zG)*idr2*idr


plt.streamplot(xG, zG, vx, vz,density=.45, arrowsize=2, broken_streamlines=True); 
spd = vx*vx + vz*vz; plt.pcolor(xG, zG, np.log(spd), cmap=plt.cm.bone); plt.colorbar()

plt.plot(0, 0, 'o', ms=32, mfc='g')
```

```python
xx=6;  x = np.linspace(-xx,xx,128); 
 
Lx, Ly = 2*np.pi, 2*np.pi
Np =64
xG, zG = np.meshgrid(np.linspace(-Lx, Lx, Np), np.linspace(-Lx, Lx, Np))
s0=1
r = np.sqrt(xG*xG + zG*zG)
idr = 1/r
idr2 = idr*idr 

ex, ez = 0, 1
eR2 = (ex*xG + ez*zG)*(ex*xG + ez*zG)*idr2

fac2S=(3*zG*zG - xG*xG)*idr2*idr2*idr

vx = s0*xG*fac2S
vz = s0*zG*fac2S


plt.streamplot(xG, zG, vx, vz,density=.83, arrowsize=1, broken_streamlines=True); 
spd = vx*vx + vz*vz; plt.pcolor(xG, zG, np.log(spd), cmap=plt.cm.bone); plt.colorbar()

plt.plot(0, 0, 'o', ms=32, mfc='g')
```
