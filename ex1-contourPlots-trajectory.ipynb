```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Np=32
x = np.linspace(0, 2*np.pi, Np)
```

```python
np.size(x)
```

```python
y = np.sin(x)
plt.plot(x, y, 'ro-')
plt.grid()
```

```python
dx = x[1] - x[0]

dydx = np.gradient(y, dx)
```

```python
plt.plot(x, dydx, 'ro-', label='numerical differential')
plt.plot(x, np.cos(x), 'b--', label='exact differential')
plt.legend()
```

## Contour plots and vector field in 2D

```python
Lx, Ly = 5, 5
xG, yG = np.meshgrid(np.linspace(-Lx, Lx, Np), np.linspace(-Lx, Lx, Np))
```

```python
V = xG*xG*yG*yG
```

```python
plt.contour(xG, yG, V, cmap='RdBu')
plt.axis('square');
```

```python
V = xG*xG*yG*yG
Fy, Fx = np.gradient(-V)
plt.quiver(xG, yG, Fx, Fy)
plt.contour(xG, yG, V, cmap='RdBu')
plt.axis('square');
```

```python
cf = plt.contourf(xG, yG, V, cmap='RdBu'); plt.colorbar()
plt.contour(cf)
plt.axis('square');
```

```python
V = xG*xG*yG*yG
Fy, Fx = np.gradient(-V)
Fcf = plt.contourf(xG, yG, V, cmap='RdBu'); plt.colorbar()
plt.quiver(xG, yG, Fx, Fy)
plt.contour(cf)
plt.axis('square');
```

```python
V = xG*xG + yG*yG
Fy, Fx = np.gradient(-V)
plt.quiver(xG, yG, Fx, Fy)
plt.contour(xG, yG, V, cmap='RdBu')
plt.axis('square');
```

```python
V = xG*xG + yG*yG
Fy, Fx = np.gradient(-V)
cf = plt.contourf(xG, yG, V, cmap='RdBu'); plt.colorbar()
plt.quiver(xG, yG, Fx, Fy)
plt.contour(cf)
plt.axis('square');
```

```python
V = (xG-yG)**2
cf = plt.contourf(xG, yG, V, cmap='RdBu'); plt.colorbar()
plt.contour(cf)
plt.axis('equal')
```

```python
Fy, Fx = np.gradient(-V)
cf = plt.contourf(xG, yG, V, cmap='RdBu'); plt.colorbar()
plt.quiver(xG, yG, Fx, Fy)
plt.contour(cf)
plt.axis('equal');
```

## Plotting 3D trajectories

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

Nt=128
t = np.linspace(0, 2*np.pi, Nt)

z0=10
X, Y, Z = np.cos(t), np.sin(t), z0

sp=ax.scatter3D(X, Y, Z, c=t)
fig.colorbar(sp);
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

Nt=10**4
t = np.linspace(0, 2*np.pi, Nt)

z0, t0 =10, 10
X, Y, Z = (t**2 - t0**2)/(t**2 + t0**2), 2*t*t0/(t**2 + t0**2), z0

sp=ax.scatter3D(X, Y, Z, c=t)
fig.colorbar(sp);
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

Nt=128
t = np.linspace(0, 3.6*np.pi, Nt)

vo=10
X, Y, Z = np.cos(t), np.sin(t), vo*t

sp=ax.scatter3D(X, Y, Z, c=t)
fig.colorbar(sp);
```

## Hyperbolic functions
* $\cosh x = \frac{e^x + e^{-x}}{2}$

* $\sinh x = \frac{e^x - e^{-x}}{2}$

* $\tanh x = \frac{e^x - e^{-x}}{e^x + e^{-x}}$

```python
x = np.linspace(-5*np.pi, 5*np.pi)

y = 0.5*(np.exp(x) + np.exp(-x))

plt.plot(x, y, label='y')
plt.plot(x, np.cosh(x), '--',  label='cosh(x)')
plt.legend()
```

```python
x = np.linspace(-5*np.pi, 5*np.pi)

y = 0.5*(np.exp(x) - np.exp(-x))

plt.plot(x, y,  label='y')
plt.plot(x, np.sinh(x), '--',  label='cosh(x)')
plt.legend()
```

```python
x = np.linspace(-5*np.pi, 5*np.pi)

y = (np.exp(x) - np.exp(-x))/((np.exp(x) + np.exp(-x)))

plt.plot(x, y,  label='y')
plt.plot(x, np.tanh(x), '--',  label='tanh(x)')
plt.legend()
```
