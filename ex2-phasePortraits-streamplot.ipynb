```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 20, "text.usetex": True,})
```

### Simple harmonic oscillator

Consider the following potential (with $K>0$)

$$U(x)=\frac12 K x^2.$$

The corresponding force field is $$F(x)= - U'(x)=-Kx$$

```python
plt.figure(figsize=(5, 10), dpi=100)

plt.subplot(211)
xx=6;  x = np.linspace(-xx,xx,128); 

plt.plot(x, x*x, lw=2, color='indianred')
plt.axvline(x=0, c="gray", label="x=0"); plt.axhline(y=0, c="gray", label="y=0")
plt.ylabel(r'$U(x)$'); 
plt.grid()
plt.subplot(212)
Lx, Ly = 2*np.pi, 2*np.pi
Np =64
xG, dot_xG = np.meshgrid(np.linspace(-Lx, Lx, Np), np.linspace(-Lx, Lx, Np))

ddot_xG = -xG

plt.axvline(x=0, c="gray", label="x=0"); plt.axhline(y=0, c="gray", label="y=0")
plt.streamplot(xG, dot_xG, dot_xG, ddot_xG,density=0.35, broken_streamlines=False, arrowsize=2); 
plt.xlabel(r'$x$'); plt.ylabel(r'$\dot x$'); 
```

### Inverted parabola

Consider the following potential (with $K>0$)
$$U(x)=-\frac12 K x^2$$

The corresponding force field is $$F(x)= - U'(x)=Kx$$

```python
plt.figure(figsize=(5, 10), dpi=100)

plt.subplot(211)
xx=6;  x = np.linspace(-xx,xx,128); 

plt.plot(x, -x*x, lw=2, color='indianred')
plt.axvline(x=0, c="gray", label="x=0")
plt.axhline(y=0, c="gray", label="y=0")
plt.ylabel(r'$U(x)$'); 
plt.grid()

plt.subplot(212)

Lx, Ly = 2*np.pi, 2*np.pi
Np = 64
xG, dot_xG = np.meshgrid(np.linspace(-Lx, Lx, Np), np.linspace(-Lx, Lx, Np))

ddot_xG = xG

plt.axvline(x=0, c="gray", label="x=0"); plt.axhline(y=0, c="gray", label="y=0")
plt.streamplot(xG, dot_xG, dot_xG, ddot_xG,density=0.35, broken_streamlines=False, arrowsize=2); 
plt.xlabel(r'$x$'); plt.ylabel(r'$\dot x$'); 
```

### Plane pendulum

The potential energy in this case is:

$$U(x)=2mgl\sin^2 (x/2)$$


The corresponding force field is 

$$F(x)= - U'(x)=-mgl\sin(x)$$


```python
plt.figure(figsize=(10, 8), dpi=100)

plt.subplot(211)
xx=4*np.pi;  x = np.linspace(-xx,xx,128); 

plt.plot(x, np.sin(x/2)**2, lw=2, color='indianred')
plt.grid(); plt.xticks(np.arange(-Lx, Lx, step=np.pi));
plt.xlim(-xx, xx)
plt.ylabel(r'$U(x)$'); 

plt.subplot(212)

Lx, Ly = 4*np.pi, 1*np.pi
Np = 64
xG, dot_xG = np.meshgrid(np.linspace(-Lx, Lx, Np), np.linspace(-Ly, Ly, Np))

ddot_xG = -np.sin(xG)

plt.axvline(x=0, c="gray", label="x=0"); plt.axhline(y=0, c="gray", label="y=0")

# plt.streamplot(xG, dot_xG, dot_xG, ddot_xG,density=0.4, broken_streamlines=False, arrowsize=2); 
plt.plot(x, 2*np.cos(x/2), lw=2, color='C0'); 
plt.plot(x, -2*np.cos(x/2), lw=2, color='C0'); 
# plt.plot(x, -2*np.cos(x/2)**2, lw=2, color='blue'); 
plt.grid(); plt.xticks(np.arange(-Lx, Lx, step=np.pi));
plt.xlabel(r'$x$'); plt.ylabel(r'$\dot x$'); plt.ylim(-3, 3)
plt.xlim(-Lx, Lx)
```

## Cubic potential

We now consider a potential of the form:

$$U (x) = x^3$$

The corresponding force field is

$$U (x) = -3x^2$$


```python
plt.figure(figsize=(5, 10), dpi=100)

plt.subplot(211)
xx=4*np.pi;  x = np.linspace(-xx,xx,128); 

plt.plot(x,  x**3, lw=2, color='indianred')
plt.axvline(x=0, c="gray", label="x=0")
plt.axhline(y=0, c="gray", label="y=0")
plt.grid()
plt.ylabel(r'$U(x)$'); 


plt.subplot(212)

Lx, Ly = 4*np.pi, 43
Np =64
xG, dot_xG = np.meshgrid(np.linspace(-Lx, Lx, Np), np.linspace(-Ly, Ly, Np))

ddot_xG =  -3*xG**2 

plt.axvline(x=0, c="gray", label="x=0"); plt.axhline(y=0, c="gray", label="y=0")

plt.streamplot(xG, dot_xG, dot_xG, ddot_xG,density=0.3, broken_streamlines=False, arrowsize=2); 
plt.xlabel(r'$x$'); plt.ylabel(r'$\dot x$'); 
```

## Hyperbolic potential

Consider a potential given as:

$$U(x)=\tanh (x)$$

The coorespodning force is 
$$F(x)= - U'(x)=\tanh^2 (x)-1$$

```python
plt.figure(figsize=(5, 10), dpi=100)

plt.subplot(211)
xx=4*np.pi;  x = np.linspace(-xx,xx,128); 

plt.plot(x, np.tanh(x), lw=2, color='indianred')
plt.axvline(x=0, c="gray", label="x=0")
plt.axhline(y=0, c="gray", label="y=0")
plt.grid()
plt.ylabel(r'$U(x)$'); 

plt.subplot(212)

Lx, Ly = 4*np.pi, 1.6*np.pi
Np =64
xG, dot_xG = np.meshgrid(np.linspace(-Lx, Lx, Np), np.linspace(-Ly, Ly, Np))

ddot_xG = np.tanh(xG)**2 -1

plt.axvline(x=0, c="gray", label="x=0"); plt.axhline(y=0, c="gray", label="y=0")

plt.streamplot(xG, dot_xG, dot_xG, ddot_xG,density=0.4, broken_streamlines=False, arrowsize=2); 
plt.xlabel(r'$x$'); plt.ylabel(r'$\dot x$'); 
```

Consider motion in a potential

$$U(x)=\frac{x^2}{(x^2 + a^2)^2}.$$

The corresponding force field is 

$$F(x)= - U'(x)= \frac{4x^3}{(x^2+a^2)^3}-\frac{2x}{(x^2 + a^2)^2}$$

```python
plt.figure(figsize=(10, 12), dpi=100)

plt.subplot(211)
xx=3*np.pi;  x = np.linspace(-xx,xx,128); 

plt.plot(x,  x*x/(x**2 + 16)**2, lw=2, color='indianred')
plt.axvline(x=0, c="gray", label="x=0")
plt.axhline(y=0, c="gray", label="y=0")
plt.grid()
plt.ylabel(r'$U(x)$'); 


plt.subplot(212)

Lx, Ly = 4*np.pi, .28
Np =100
xG, dot_xG = np.meshgrid(np.linspace(-Lx, Lx, Np), np.linspace(-Ly, Ly, Np))

ddot_xG =  3*xG**3/(xG**2 + 16)**3 - 2*xG/(xG**2 + 16)**2

plt.axvline(x=0, c="gray", label="x=0"); plt.axhline(y=0, c="gray", label="y=0")

plt.streamplot(xG, dot_xG, dot_xG, ddot_xG,density=0.75, broken_streamlines=False, arrowsize=2); 
plt.xlabel(r'$x$'); plt.ylabel(r'$\dot x$'); 
```

## Lennard-Jones Potential

```python
plt.figure(figsize=(6, 12), dpi=100)

plt.subplot(211)
xx=3.5;  x = np.linspace(0.5,xx,256); 

plt.plot(x,  (1/x)**12 -  2*(1/x)**6, lw=2, color='indianred')
plt.axvline(x=0, c="gray", label="x=0")
plt.axhline(y=0, c="gray", label="y=0")
plt.grid(); plt.ylim (-1,1); plt.xlim(0.5, 2.5)

plt.ylabel(r'$U(x)$'); 


plt.subplot(212)

Lx, Ly = 3.5, 1.25
Np =100
xG, dot_xG = np.meshgrid(np.linspace(0.5, Lx, Np), np.linspace(-Ly, Ly, Np))
ddot_xG =  (1/xG)**13 -  (1/xG)**7

plt.axvline(x=0, c="gray", label="x=0"); plt.axhline(y=0, c="gray", label="y=0")

plt.streamplot(xG, dot_xG, dot_xG, ddot_xG,density=.5, broken_streamlines=False, arrowsize=2); 
plt.xlabel(r'$x$'); plt.ylabel(r'$\dot x$'); plt.xlim(0.5, 2.5);
```

## Double well potential

```python
plt.figure(figsize=(6, 12), dpi=100)

plt.subplot(211)
xx=1.5;  x = np.linspace(-xx,xx,256); 

plt.plot(x,  -x**2 + x**4, lw=2, color='indianred')
plt.axvline(x=0, c="gray", label="x=0")
plt.axhline(y=0, c="gray", label="y=0")
plt.grid(); plt.ylim (-0.25,1); #plt.xlim(0.5, 2.5)

plt.ylabel(r'$U(x)$'); 


plt.subplot(212)

Lx, Ly = 1.5, 1.25
Np =100
xG, dot_xG = np.meshgrid(np.linspace(-Lx, Lx, Np), np.linspace(-Ly, Ly, Np))
ddot_xG =   xG - xG**3

plt.axvline(x=0, c="gray", label="x=0"); plt.axhline(y=0, c="gray", label="y=0")

plt.streamplot(xG, dot_xG, dot_xG, ddot_xG,density=.5, broken_streamlines=False, arrowsize=2); 
plt.xlabel(r'$x$'); plt.ylabel(r'$\dot x$'); #plt.xlim(0.5, 2.5);
```
