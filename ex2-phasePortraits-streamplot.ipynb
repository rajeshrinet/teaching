```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 12})
```

## Hyperbolic functions
* $\cosh x = \frac{e^x + e^{-x}}{2}$

* $\sinh x = \frac{e^x - e^{-x}}{2}$

* $\tanh x = \frac{e^x - e^{-x}}{e^x + e^{-x}}$

```python
x = np.linspace(-5*np.pi, 5*np.pi)

y = 0.5*(np.exp(x) + np.exp(-x))

plt.plot(x, y, label='y')
plt.plot(x, np.cosh(x), '--',  label='cosh(x)')
plt.legend()
```

```python
x = np.linspace(-5*np.pi, 5*np.pi)

y = 0.5*(np.exp(x) - np.exp(-x))

plt.plot(x, y,  label='y')
plt.plot(x, np.sinh(x), '--',  label='cosh(x)')
plt.legend()
```

```python
x = np.linspace(-5*np.pi, 5*np.pi)

y = (np.exp(x) - np.exp(-x))/((np.exp(x) + np.exp(-x)))

plt.plot(x, y,  label='y')
plt.plot(x, np.tanh(x), '--',  label='tanh(x)')
plt.legend()
```

```python

plt.plot(x, y , label='Potential')
plt.plot(x, -1+np.tanh(x)**2, '--', label='Force')
plt.legend()
```

## Phase diagrams

### Simple harmonic oscillator


$U(x)=\frac12 K x^2$.

The coorespodning force is $F(x)= - U'(x)=-Kx$

The expresion for energy is

$E = \frac12 m \dot x^2 + \frac12 K x^2$

This implies that:

$\dot x = \sqrt{2(E- U)/m}$


```python
xx=1.42;  x = np.linspace(-xx,xx,128); 

plt.figure(figsize=(4, 4), dpi=80)

plt.plot(x, x*x)
```

```python
plt.figure(figsize=(6, 6), dpi=100)
Lx, Ly = 2*np.pi, 2*np.pi
Np = 32
xG, xdotG = np.meshgrid(np.linspace(-Lx, Lx, Np), np.linspace(-Lx, Lx, Np))

xddotG = -xG

plt.streamplot(xG, xdotG, xdotG, xddotG)
```

### Inverted parabola


$U(x)=-\frac12 K x^2$.

The coorespodning force is $F(x)= - U'(x)=Kx$

The expresion for energy is

$E = \frac12 m \dot x^2 - \frac12 K x^2$

This implies that:

$\dot x = \sqrt{2(E- U)/m}$


```python
xx=1.42;  x = np.linspace(-xx,xx,128); 

plt.figure(figsize=(4, 4), dpi=80)

plt.plot(x, -x*x)
```

```python
plt.figure(figsize=(6, 6), dpi=100)

Lx, Ly = 2*np.pi, 2*np.pi
Np = 32
xG, xdotG = np.meshgrid(np.linspace(-Lx, Lx, Np), np.linspace(-Lx, Lx, Np))

xddotG = xG

plt.streamplot(xG, xdotG, xdotG, xddotG);
```

### Plane pendulum

$U(x)=2mgl\sin^2 (x/2)$.

The coorespodning force is $F(x)= - U'(x)=-mgl\sin(x)$

The expresion for energy is

$E = \frac12 m \dot x^2 + 2mgl\sin\theta$

This implies that:

$\dot x = \sqrt{2(E- U)/m}$


```python
xx=8.42;  x = np.linspace(-xx,xx,128); 

plt.figure(figsize=(8, 4), dpi=80)

plt.plot(x, np.sin(x/2)**2)
```

```python
plt.figure(figsize=(12, 4), dpi=100)
Lx, Ly = 3*np.pi, 1.2*np.pi
Np = 32
xG, xdotG = np.meshgrid(np.linspace(-Lx, Lx, Np), np.linspace(-Ly, Ly, Np))

xddotG = -np.sin(xG)

plt.streamplot(xG, xdotG, xdotG, xddotG)
```

## Consider motion in a potential

$U(x)=\tanh (x)$.

The coorespodning force is $F(x)= - U'(x)=\tanh^2 (x)-1$

```python
xx=5;  x = np.linspace(-xx,xx,128); 

plt.figure(figsize=(4, 4), dpi=80)

plt.plot(x, np.tanh(x))
```

```python
plt.figure(figsize=(8, 4), dpi=100)

Lx, Ly = np.pi, np.pi
Np = 32
xG, xdotG = np.meshgrid(np.linspace(-Lx, Lx, Np), np.linspace(-Ly, Ly, Np))

xddotG = np.tanh(xG)**2 -1

plt.streamplot(xG, xdotG, xdotG, xddotG)
```

## Consider motion in a potential

$U(x)=\frac{x^2}{(x^2 + a^2)^2}$.

The corresponding force is $F(x)= - U'(x)= \frac{4x^3}{(x^2+a^2)^3}-\frac{2x}{(x^2 + a^2)^2}$

```python
xx=15;  x = np.linspace(-xx,xx,128); 

plt.figure(figsize=(4, 4), dpi=80)

plt.plot(x, x*x/(x**2 + 16)**2)
```

```python
plt.figure(figsize=(12, 4), dpi=100)

Lx, Ly = 4*np.pi, .28
Np = 64
xG, xdotG = np.meshgrid(np.linspace(-Lx, Lx, Np), np.linspace(-Ly, Ly, Np))

xddotG =  4*xG**3/(xG**2 + 16)**3 - 2*xG/(xG**2 + 16)**2

plt.streamplot(xG, xdotG, xdotG, xddotG)
```
