```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 12})
```

### Simple harmonic oscillator


$U(x)=\frac12 K x^2$.

The coorespodning force is $F(x)= - U'(x)=-Kx$

The expresion for energy is

$E = \frac12 m \dot x^2 + \frac12 K x^2$

This implies that:

$\dot x = \sqrt{2(E- U)/m}$


```python
plt.figure(figsize=(8, 12), dpi=100)

plt.subplot(211)
xx=6;  x = np.linspace(-xx,xx,128); 

plt.plot(x, x*x, lw=2, color='indianred')
plt.axvline(x=0, c="gray", label="x=0")
plt.axhline(y=0, c="gray", label="y=0")
plt.grid()
plt.subplot(212)
Lx, Ly = 2*np.pi, 2*np.pi
Np =64
xG, dot_xG = np.meshgrid(np.linspace(-Lx, Lx, Np), np.linspace(-Lx, Lx, Np))

ddot_xG = -xG

plt.streamplot(xG, dot_xG, dot_xG, ddot_xG,density=0.75, broken_streamlines=False, arrowsize=2); 
```

### Inverted parabola


$U(x)=-\frac12 K x^2$.

The coorespodning force is $F(x)= - U'(x)=Kx$

The expresion for energy is

$E = \frac12 m \dot x^2 - \frac12 K x^2$

This implies that:

$\dot x = \sqrt{2(E- U)/m}$


```python
plt.figure(figsize=(8, 12), dpi=100)

plt.subplot(211)
xx=6;  x = np.linspace(-xx,xx,128); 

plt.plot(x, -x*x, lw=2, color='indianred')
plt.axvline(x=0, c="gray", label="x=0")
plt.axhline(y=0, c="gray", label="y=0")
plt.grid()

plt.subplot(212)

Lx, Ly = 2*np.pi, 2*np.pi
Np = 64
xG, dot_xG = np.meshgrid(np.linspace(-Lx, Lx, Np), np.linspace(-Lx, Lx, Np))

ddot_xG = xG

plt.streamplot(xG, dot_xG, dot_xG, ddot_xG,density=0.5, broken_streamlines=False, arrowsize=2); 
```

### Plane pendulum

$U(x)=2mgl\sin^2 (x/2)$.

The coorespodning force is $F(x)= - U'(x)=-mgl\sin(x)$

The expresion for energy is

$E = \frac12 m \dot x^2 + 2mgl\sin\theta$

This implies that:

$\dot x = \sqrt{2(E- U)/m}$


```python
plt.figure(figsize=(14, 10), dpi=100)

plt.subplot(211)
xx=3*np.pi;  x = np.linspace(-xx,xx,128); 

plt.plot(x, np.sin(x/2)**2, lw=2, color='indianred')
plt.grid(); plt.xticks(np.arange(-Lx, Lx, step=np.pi));
plt.xlim(-xx, xx)

plt.subplot(212)

Lx, Ly = 3*np.pi, 1*np.pi
Np = 64
xG, dot_xG = np.meshgrid(np.linspace(-Lx, Lx, Np), np.linspace(-Ly, Ly, Np))

ddot_xG = -np.sin(xG)

plt.streamplot(xG, dot_xG, dot_xG, ddot_xG,density=0.75, broken_streamlines=False, arrowsize=2); 
plt.grid(); plt.xticks(np.arange(-Lx, Lx, step=np.pi));
```

## Hyperbolic potential

$U(x)=\tanh (x)$.

The coorespodning force is $F(x)= - U'(x)=\tanh^2 (x)-1$

```python
plt.figure(figsize=(15, 12), dpi=100)

plt.subplot(211)
xx=4*np.pi;  x = np.linspace(-xx,xx,128); 

plt.plot(x, np.tanh(x), lw=2, color='indianred')
plt.axvline(x=0, c="gray", label="x=0")
plt.axhline(y=0, c="gray", label="y=0")
plt.grid()

plt.subplot(212)

Lx, Ly = 2*np.pi, 1*np.pi
Np =64
xG, dot_xG = np.meshgrid(np.linspace(-Lx, Lx, Np), np.linspace(-Ly, Ly, Np))

ddot_xG = np.tanh(xG)**2 -1

plt.streamplot(xG, dot_xG, dot_xG, ddot_xG,density=0.75, broken_streamlines=False, arrowsize=2); 
```

## Consider motion in a potential

$U(x)=\frac{x^2}{(x^2 + a^2)^2}$.

The corresponding force is $F(x)= - U'(x)= \frac{4x^3}{(x^2+a^2)^3}-\frac{2x}{(x^2 + a^2)^2}$

```python
plt.figure(figsize=(15, 12), dpi=100)

plt.subplot(211)
xx=4*np.pi;  x = np.linspace(-xx,xx,128); 

plt.plot(x,  x*x/(x**2 + 16)**2, lw=2, color='indianred')
plt.axvline(x=0, c="gray", label="x=0")
plt.axhline(y=0, c="gray", label="y=0")
plt.grid()

plt.subplot(212)

Lx, Ly = 4*np.pi, .325
Np =64
xG, dot_xG = np.meshgrid(np.linspace(-Lx, Lx, Np), np.linspace(-Ly, Ly, Np))

ddot_xG =  4*xG**3/(xG**2 + 16)**3 - 2*xG/(xG**2 + 16)**2

plt.streamplot(xG, dot_xG, dot_xG, ddot_xG,density=0.75, broken_streamlines=False, arrowsize=2); 
```
