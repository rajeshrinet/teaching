## The orbit equation in the central-force problem

$$
\frac1r = \frac1r_0\left(1 + e \cos\phi\right)
$$

Rearranging, we get

$$
r =   \frac{r_0}{\left(1 + e \cos\phi\right)}
$$

```python
N=512
r0 = .1
e = .8

phi = np.linspace(0, 2*np.pi, N)
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.plot(phi, r0/(1+e*np.cos(phi)), 'r')

ax.set_rmax(1.12*r0/(1-e)); 
ax.grid(True)
```

## Motion in -$\frac{1}{r^3}$ potential

### Archimedean spiral

$${\displaystyle r=a+b\cdot \phi }$$

```python
N=512
a = .0
b = 1

phi = np.linspace(0, 4*np.pi, N)
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.plot(phi, a + b*phi, 'r')
ax.plot(phi[0], (a + b*phi)[0], 'o')

# ax.set_rmax(2); 
ax.grid(True)
```

### Cote's spiral

$$\frac1r= C\cosh \phi $$

```python
N=512
C = 1
alpha=0.3
phi = np.linspace(0, 5*np.pi, N)
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.plot(phi, 1/(C*np.cosh(alpha*phi)), 'steelblue',lw=2,)
ax.plot(phi[0], (1/(C*np.cosh(alpha*phi)))[0], 'ro')


# ax.set_rmax(2); 
ax.grid(True)
```

```python
N=512
A = 1
B = .3

phi = np.linspace(0, 4*np.pi, N)
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.plot(phi, 1/(A + B*(phi)), 'steelblue',lw=2,)

ax.plot(phi[0], (A + B*phi)[0], 'ro')

# ax.set_rmax(2); 
ax.grid(True)
```

```python
N=512
C = 11
beta=.22
phi = np.linspace(0*np.pi, 2*np.pi, N)
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.plot(phi, 1/(C*np.cos(beta*phi)),  'steelblue',lw=2,)
ax.plot(phi[0], (1/(C*np.cos(beta*(phi-0.0*np.pi))))[0], 'ro')


# ax.set_rmax(2); 
ax.grid(True)
```
